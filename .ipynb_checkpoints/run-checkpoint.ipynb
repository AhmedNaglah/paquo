```python
import shapely
from shapely.geometry import Polygon

ps = []
for annotation in image.hierarchy.annotations:
    if annotation.path_class.name=="tbm_add":
        ps.append(annotation.roi)

print(len(ps))

cnts = []

for poly in ps:
    if poly.type == 'Polygon':
        xx, yy = poly.exterior.xy
        points = [[x, y] for x, y in  zip(xx, yy)]
        cnts.append(points)
    elif poly.type == 'MultiPolygon':
        xys = [poly_item.exterior.xy for poly_item in poly.geoms]
        for xy in xys:
            xx, yy = xy
            points = [[x, y] for x, y in  zip(xx, yy)]
            cnts.append(points)     

print(len(cnts))
```

```python
def intersection(x1,x2,x3,x4,y1,y2,y3,y4):
    d = (x1-x2)*(y3-y4) - (y1-y2)*(x3-x4)
    if d:
        xs = ((x1*y2-y1*x2)*(x3-x4) - (x1-x2)*(x3*y4-y3*x4)) / d
        ys = ((x1*y2-y1*x2)*(y3-y4) - (y1-y2)*(x3*y4-y3*x4)) / d
        if (xs >= min(x1,x2) and xs <= max(x1,x2) and
            xs >= min(x3,x4) and xs <= max(x3,x4)):
            return xs, ys
x = xx
y = yy
xs, ys = [], []
for i in range(len(x)-1):
    for j in range(i-1):
        if xs_ys := intersection(x[i],x[i+1],x[j],x[j+1],y[i],y[i+1],y[j],y[j+1]):
            xs.append(xs_ys[0])
            ys.append(xs_ys[1])
```
